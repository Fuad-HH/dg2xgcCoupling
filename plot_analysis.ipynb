### Plot the radial basis function

```python
import numpy as np
```

```python
def radial_basis_function(r, rho):
    if (abs(r) > rho):
        return 0
    nr = abs(r)/rho
    phi = (5*nr**5 + 30*nr**4 + 72*nr**3 + 82*nr**2 + 36*nr + 6)*(1-nr)**6
    return phi
```

```python
# plot the radial basis function for different values of rho
rho = [5., 8.]
r = np.linspace(-10, 10, 1000)
phi_array = np.zeros((len(rho), len(r)))

for i in range(len(rho)):
    for j in range(len(r)):
        phi_array[i,j] = radial_basis_function(r[j], rho[i])
```

```python
# plot 
plt.plot(r, phi_array[0], 'r-', label=r'$\rho = 5$')
plt.plot(r, phi_array[1], 'b--', label=r'$\rho = 8$')
# draw a vertical line at r=5 and r=8 as well as negative of these values
plt.axvline(x=5, color='r', linestyle="dotted")
plt.axvline(x=-5, color='r', linestyle="dotted")
plt.axvline(x=8, color='b', linestyle="dotted")
plt.axvline(x=-8, color='b', linestyle="dotted")
plt.xlabel('r')
plt.ylabel(r"$\phi (r)$")
plt.legend()
plt.title('Radial Basis Function')
plt.show()
```
